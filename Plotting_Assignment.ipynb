# Histogram 1 (interpretation)
The distribution shows a skew toward lower CO emissions. Most vehicles emit between 300-450 g/mile, but there's a long tail of higher emitters. This suggests that while many vehicles are relatively efficient, a portion of the fleet still emits a lot of CO.

```python
# Histogram for Fuel Efficiency Score
plt.figure(figsize=(10, 5))
plt.hist(df['feScore'], bins=15, color='skyblue', edgecolor='black')
plt.title('Fuel Efficiency Score Distribution')
plt.xlabel('Fuel Efficiency Score')
plt.ylabel('Frequency')
plt.show()
```

# Histogram 2 (Interpretation)
Fuel Efficiency Score is mostly concentrated between 4 and 7, meaning the bulk of vehicles have moderate efficiency.

```python
# Histogram for Combined Fuel Economy
plt.figure(figsize=(10, 5))
plt.hist(df['comb'], bins=15, color='gold', edgecolor='black')
plt.title('Combined Fuel Economy Distribution')
plt.xlabel('Combined MPG')
plt.ylabel('Frequency')
plt.show()
```

# Histogram 3(interpritation)
Combined MPG ranges widely, with many vehicles around 20-30 MPG, showing a good diversity in vehicle fuel economy.

```python
plt.figure(figsize=(12, 8))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, cmap='coolwarm', annot=False, linewidths=0.5)
plt.title('Correlation Heatmap of Vehicle Features')
plt.xticks(rotation=45)
plt.show()
```

#Interpretation of the Heatmap:

co2 is negatively correlated with feScore, ghgScore, and comb, which is expected: better MPG and fuel scores = lower CO.

displ (engine displacement) and cylinders have a positive correlation with co2, since larger engines typically emit more.

pv2 and pv4 have very little correlation with ghgScore, so they might not be that useful in predicting it.
