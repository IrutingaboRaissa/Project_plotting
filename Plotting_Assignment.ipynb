# Histogram 1 (interpretation)
The histogram shows a skew toward lower CO2 emissions. Most vehicles emit between 300-450, but there are higher emitters. This suggests that while many vehicles are different, a number of them still emits a lot of CO2.

```python
# Histogram for Fuel Efficiency Score
plt.figure(figsize=(10, 10))
plt.hist(df['feScore'], bins=10, color='skyblue', edgecolor='Red')
plt.title('Fuel Efficiency Score Distribution')
plt.xlabel('Fuel Efficiency Score')
plt.ylabel('Frequency')
plt.show()
```

# Histogram 2 (Interpretation)
Fuel Efficiency Score is mostly concentrated between 4 and 8, meaning most vehicles have average effectiveness.

```python
# Histogram for Combined Fuel Economy
plt.figure(figsize=(10, 8))
plt.hist(df['comb'], bins=18, color='gold', edgecolor='black')
plt.title('Combined Fuel Economy Distribution')
plt.xlabel('Comb')
plt.ylabel('Frequency')
plt.show()
```

# Histogram 3(interpritation)
Comb fuel efficiency ranges widely, with many vehicles around 20–30, showing diversity in vehicle performance.

```python
plt.figure(figsize=(12, 10))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, cmap='coolwarm', annot=True, linewidths=0.5)
plt.title('Correlation Heatmap of Vehicle Features')
plt.xticks(rotation=45)
plt.show()
```

# Interpretation of the Heatmap:

CO2 is negatively correlated with feScore, ghgScore, and comb, which is expected: better fuel scores = lower CO2.

displ and cylinders have a positive correlation with CO2, since larger engines emit more.
